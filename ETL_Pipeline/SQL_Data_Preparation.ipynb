# Raw Data Cleaning and Preparation for SQL Database

## Import required packages and raw dataset

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/raw_marketing_campaign_events.csv')
df
```

```python
df.info()
```

## Data Cleaning

```python
# remove the extra whitespace in campaign_name column
print(df['campaign_name'].value_counts())
df['campaign_name']=df['campaign_name'].str.strip()
print(df['campaign_name'].value_counts())
```

```python
# Use the the consistent format in campaign type
print(df['campaign_type'].value_counts())
df['campaign_type']=df['campaign_type'].str.capitalize()
print(df['campaign_type'].value_counts())
```

```python
#remove white space around or in the email address and convert invalid or empty record to missing or invlaid email
print(df['customer_email'].value_counts())
df['customer_email']=(df['customer_email'].str.strip()).str.lower()
df['customer_email']=df['customer_email'].str.replace(' ','')
df['customer_email'] = df['customer_email'].fillna('Missing Email')
mask=df['customer_email'].str.find('@')==-1
df.loc[mask,'customer_email']='Missing or Invalid Email'
#df['customer_email']
```

```python
#fix the date format
df['sent_date']=pd.to_datetime(df['sent_date'],format='mixed')
```

```python
# use the consistent formart, Y for yes and N for No
df['opened'].value_counts()
mask=df['opened'].isin(['Yes','true','TRUE'])
df.loc[mask,'opened']='Y'
mask=df['opened'].isin(['No','0'])
df.loc[mask,'opened']='N'
print(df['opened'].value_counts())

df['clicked'].value_counts()
mask=df['clicked'].isin(['Yes','TRUE'])
df.loc[mask,'clicked']='Y'
mask=df['clicked'].isin(['No','0','false'])
df.loc[mask,'clicked']='N'
print(df['clicked'].value_counts())

df['converted'].value_counts()
mask=df['converted'].isin(['No','false','0'])
df.loc[mask,'converted']='N'
mask=df['converted'].isin(['Yes','1','TRUE'])
df.loc[mask,'converted']='Y'
print(df['converted'].value_counts())

df['unsubscribe'].value_counts()
mask=df['unsubscribe'].isin(['No','false','0','FALSE'])
df.loc[mask,'unsubscribe']='N'
print(df['unsubscribe'].value_counts())
```

```python
# remove the dollar sign from the dataset and keep the missing value as it means we don't have any value
df['conversion_amount'].value_counts()
df['conversion_amount']=df['conversion_amount'].str.replace('$','')
df['conversion_amount'].value_counts()
df['conversion_amount']=pd.to_numeric(df['conversion_amount'])
```

```python
# Remove duplicates
df.info()
df.drop_duplicates(inplace=True)
#df.info()
#df
df2=pd.read_csv('/content/customer_202606171145.csv')
```

# Prepare the Data for importing into SQL Database

```python
#Split the data and generate the primary key for each entity
complete_data = pd.merge(
    df,
    df2,
    left_on='customer_email',
    right_on='email_address',
    how='left'
)

columns=list(df.columns)
columns.append('customer_id')
complete_data=complete_data[columns]
#complete_data

actual_data=complete_data[complete_data['customer_id'].isna()==False]
rejected_data=complete_data[complete_data['customer_id'].isna()]

campaign=actual_data[['campaign_name','campaign_type']]
campaign=campaign.drop_duplicates()
campaign = campaign.reset_index(drop=True)
campaign['campaign_id'] = campaign.index + 1

data=pd.merge(
    campaign,
    actual_data,
    left_on='campaign_name',
    right_on='campaign_name',
    how='inner'
)
data['campaign_type']=data['campaign_type_y']
data=data.drop(columns=['campaign_type_x','campaign_type_y'])
data['customer_campaign_id']=data.index+1
campaign=data[['campaign_id','campaign_name','campaign_type']]
campaign.drop_duplicates(inplace=True)
customer_campaign=data
customer_campaign=customer_campaign.drop(columns=['campaign_name','campaign_type','customer_email'])
campaign.to_csv('/content/campaign.csv',index=False)
customer_campaign.to_csv('/content/customer_campaign.csv',index=False)
rejected_data.to_csv('/content/rejected_data.csv',index=False)
```
